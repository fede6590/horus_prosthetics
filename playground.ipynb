```python
import cv2
import numpy as np
```

```python
# Load an image
image = cv2.imread('img.png')
```

```python

# Detect keypoints and descriptors
detector = cv2.ORB_create()
keypoints, descriptors = detector.detectAndCompute(image, None)
```

```python

# Measure the angle between keypoints
angle = np.arctan2(keypoints[1].pt[1] - keypoints[0].pt[1], keypoints[1].pt[0] - keypoints[0].pt[0])

# Draw lines to visualize the keypoints
cv2.drawKeypoints(image, keypoints, image)

# Display the image with keypoints and angle measurement
cv2.imshow('Image with Keypoints', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
